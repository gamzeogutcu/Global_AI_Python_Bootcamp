```python
import random

def rock_paper_scissors_GAMZE_OGUTCU():
    print("Welcome to one of the most famous games in the world! Rock, Paper, Scissors!")
    print("Rules:")
    print("1. Choose one of the following: Rock, Paper, or Scissors to play, or if you do not want to play, write 'exit'.")
    print("2. The first to win two rounds wins the game.")
    print("3. Have fun!\n")

    options = ["rock", "paper", "scissors"]

    total_games = 0  # Initialize total_games outside of the loop

    while True:
        total_games += 1
        player_wins = 0
        computer_wins = 0
        round_count = 0

        print("\n##########################################################################")
        print(f"Game {total_games} begins!")

        while player_wins < 2 and computer_wins < 2:
            round_count += 1
            print(f"\n##########################################################################")
            print(f"Game {total_games}, Round {round_count}!")

            choice = input("Choose Rock, Paper, Scissors or exit: ").lower()

            if choice == "exit":
                print("Exiting the game. Thanks for playing!")
                return

            if choice not in options:
                print("Invalid choice! Please select rock, paper, or scissors.")
                continue

            computer_choice = random.choice(options)
            print(f"Computer chose: {computer_choice}")

            if choice == computer_choice:
                print("It's a tie!")
            elif (choice == "rock" and computer_choice == "scissors") or \
                 (choice == "paper" and computer_choice == "rock") or \
                 (choice == "scissors" and computer_choice == "paper"):
                print("You win this round!")
                player_wins += 1
            else:
                print("Computer wins this round!")
                computer_wins += 1

            print(f"Your wins: {player_wins}, Computer's wins: {computer_wins}")
            print("##########################################################################")

        if player_wins == 2:
            print(f"Congratulations! You won Game {total_games}.")
        else:
            print(f"Computer won Game {total_games}.")

        # Ask the user if they want to play another game
        play_again = input("Do you want to play another game? (yes/no): ").lower()

        if play_again == "no":
            print("Game over. Thanks for playing!")
            break

        # Determine if the computer wants to play again
        computer_play_again = random.choice(["yes", "no"])
        print(f"Computer wants to play again: {computer_play_again.capitalize()}")

        if computer_play_again == "no":
            print("The computer does not want to play again. Game over. Thanks for playing!")
            break
```

```python
rock_paper_scissors_GAMZE_OGUTCU()
```
